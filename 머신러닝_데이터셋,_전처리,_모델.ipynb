```python
import pandas as pd
import io
from google.colab import files

def analyze_dataframe(df, name):
    # 1. 데이터의 크기 (행, 열)
    print(f"\n[1] 데이터 크기 (Shape):")
    print(f"  - 행 (Rows): {df.shape[0]}")
    print(f"  - 열 (Columns): {df.shape[1]}")

    # 2. 데이터 정보 (컬럼명, 데이터 타입, non-null 개수)
    print(f"\n[2] 데이터 정보 (Info):")
    # info() 결과를 문자열로 받아와서 출력
    buffer = io.StringIO()
    df.info(buf=buffer)
    info_str = buffer.getvalue()
    print(info_str)

    # 3. 결측치(Missing Values) 개수
    print(f"[3] 컬럼별 결측치 개수:")
    missing_values = df.isnull().sum()
    # 결측치가 있는 컬럼만 출력
    if missing_values.sum() == 0:
        print("  - 결측치가 없습니다.")
    else:
        print(missing_values[missing_values > 0])

    # 4. 데이터 미리보기 (상위 5개)
    display(df.head())
    print("1. 서울특별시_관악구_가로휴지통_현황_20230905.csv")
    print("2. 서울시_상권분석서비스(영역-상권).csv")
    print("3. 서울시_상권분석서비스(길단위인구-상권).csv")


uploaded = files.upload()

# --- 각 데이터셋 불러와서 분석 함수 실행 ---
for filename, content in uploaded.items():
    df = pd.read_csv(io.BytesIO(content), encoding='cp949')
    analyze_dataframe(df, filename)
```

```python
# =================================================================
# 관악구 폐기물 관리 프로젝트 1단계: 데이터 전처리 및 위치 정보 구체화
# 환경: Google Colab
# 목표: 1. 주소 데이터를 위도/경도 좌표로 변환 (지오코딩)
#       2. TM좌표계 데이터를 WGS84(위도/경도) 좌표계로 변환
#       3. 전처리된 데이터들을 지도에 시각화하여 확인
# =================================================================


# --- 0. 기본 환경 설정 ---
print("--- 0. 기본 환경 설정 시작 ---")
!pip install geopandas pyproj folium > /dev/null
print("  - geopandas, pyproj, folium 라이브러리 설치 완료.")

import pandas as pd
import numpy as np
import folium
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderUnavailable
from pyproj import Transformer
from tqdm import tqdm
import time
import warnings
from google.colab import files # 파일 업로드를 위한 라이브러리
import io # 업로드된 파일을 메모리에서 읽기 위한 라이브러리
import re # 주소 정제를 위한 정규표현식 라이브러리

warnings.filterwarnings('ignore')
print("  - 라이브러리 임포트 완료.\n")


# --- 1. 데이터 파일 업로드 및 불러오기 (더욱 개선된 방식) ---
print("--- 1. 데이터 파일 업로드 ---")
print("1. 서울특별시_관악구_가로휴지통_현황_20230905.csv")
print("2. 서울시_상권분석서비스(영역-상권).csv")

try:
    # Colab의 파일 업로드 기능 실행
    uploaded = files.upload()

    # 업로드된 파일들을 동적으로 식별하여 데이터프레임으로 변환
    trash_bins_df = None
    commercial_df = None

    # 업로드된 파일들의 이름을 확인하고, 키워드를 통해 적절한 데이터프레임에 할당
    for filename in uploaded.keys():
        if '가로휴지통' in filename:
            print(f"'{filename}' 파일을 가로휴지통 데이터로 인식합니다.")
            trash_bins_df = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='cp949')
        elif '상권분석' in filename:
            print(f"'{filename}' 파일을 상권분석 데이터로 인식합니다.")
            commercial_df = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='cp949')

    # 두 파일이 모두 성공적으로 로드되었는지 확인
    if trash_bins_df is None or commercial_df is None:
        raise ValueError("필요한 CSV 파일 2개(가로휴지통, 상권분석)가 모두 업로드되지 않았거나, 파일 이름에 키워드가 없어 식별할 수 없습니다.")

    print("\n  - CSV 파일 2개 업로드 및 불러오기 성공!")
    print(f"  - 가로휴지통 데이터: {len(trash_bins_df)} 행")
    print(f"  - 상권 영역 데이터: {len(commercial_df)} 행\n")

except ValueError as e:
    print(f"\n[오류] {e}")
    # 파일이 없으면 이후 코드 실행이 불가하므로 중단
    raise


# --- 2. 주소의 좌표 변환 (지오코딩) ---
# '서울특별시 관악구_가로휴지통 현황' 데이터의 주소를 위도/경도 좌표로 변환합니다.
# '건물명', '부근' 등 불필요한 정보를 제거하고 표준 주소 형식에 가깝게 만듭니다.
def create_refined_address(row):
    doro = str(row['도로명'])
    seolchi = str(row['설치위치'])

    # 패턴 1: '도로명'이 이미 'OO동 123-4' 같은 지번 주소 형태일 경우
    # 예: '관악구 신림동 147-2' -> 이 주소를 그대로 사용
    if '동 ' in doro and any(char.isdigit() for char in doro):
        return f"서울특별시 {doro}"

    # 패턴 2: '도로명'이 'OO로 123' 또는 'OO길 12' 같은 도로명 주소 형태일 경우
    # '설치위치'는 무시하고 '도로명'을 정제하여 사용
    # 예: 도로명 '관악로1' -> '관악로 1', '남부순환로 1695' -> '남부순환로 1695'
    doro_cleaned = re.sub(r'([가-힣])(\d)', r'\1 \2', doro).strip() # '관악로1' -> '관악로 1'
    if ('로 ' in doro_cleaned or '길 ' in doro_cleaned) and any(char.isdigit() for char in doro_cleaned):
        # 도로명 주소 패턴 뒤에 붙은 불필요한 설명(예: 아파트 이름)을 제거
        match = re.search(r'([가-힣\d\s]+(로|길))\s*([\d-]+)', doro_cleaned)
        if match:
             return f"서울특별시 관악구 {match.group(0)}"
        return f"서울특별시 관악구 {doro_cleaned}"

    # 패턴 3: '도로명'이 지명이고 '설치위치'에 상세 정보가 있을 경우
    # 예: 도로명 '신림역', 설치위치 '1번출구 앞'
    # -> '서울특별시 관악구 신림역' 까지만 사용. '1번출구 앞' 같은 세부 정보는 지오코딩에 방해가 될 수 있음.
    return f"서울특별시 관악구 {doro}"

trash_bins_df['refined_address'] = trash_bins_df.apply(create_refined_address, axis=1)


# geopy 라이브러리 및 Nominatim 서비스 설정
geolocator = Nominatim(user_agent="gwanak-waste-optimization-project-v3")

latitudes = []
longitudes = []
geocoding_failures = 0

# tqdm을 사용하여 진행 상황을 시각적으로 표시합니다.
for address in tqdm(trash_bins_df['refined_address'], desc="지오코딩 진행률"):
    try:
        location = geolocator.geocode(address, timeout=10)
        if location:
            latitudes.append(location.latitude)
            longitudes.append(location.longitude)
        else:
            latitudes.append(np.nan) # 주소를 찾지 못한 경우 결측치로 처리
            longitudes.append(np.nan)
            geocoding_failures += 1

        # Nominatim의 과도한 사용 방지 정책을 준수하기 위해 요청 사이에 1초의 지연을 줍니다.
        time.sleep(1)

    except (GeocoderTimedOut, GeocoderUnavailable):
        print(f"서비스 시간 초과 또는 오류 발생: {address}")
        latitudes.append(np.nan)
        longitudes.append(np.nan)
        geocoding_failures += 1
        time.sleep(5) # 서비스 불안정 시 잠시 대기

trash_bins_df['위도'] = latitudes
trash_bins_df['경도'] = longitudes

print(f"\n  - 지오코딩 완료! 총 {len(trash_bins_df)}개 주소 중 {len(trash_bins_df) - geocoding_failures}개 성공.")
# 결측치가 있는 행의 개수 확인
print(f"  - 좌표 변환 실패 (결측치): {trash_bins_df['위도'].isnull().sum()} 개")
print("  - 지오코딩 결과 (상위 5개):")
print(trash_bins_df[['refined_address', '위도', '경도']].head())
print("\n")


# --- 3. 좌표계 변환 ---
# '서울시 상권분석서비스(영역-상권)' 데이터의 TM 좌표를 위도/경도(WGS84)로 변환합니다.
# (보고서 계획 II-B-1 해당)
print("--- 3. 상권 영역 좌표계 변환 시작 ---")

# 전체 상권 데이터에서 관악구 데이터만 필터링합니다.
commercial_gwanak_df = commercial_df[commercial_df['자치구_코드_명'] == '관악구'].copy()

# 좌표 변환기 설정 (EPSG:5181 -> EPSG:4326)
transformer = Transformer.from_crs("epsg:5181", "epsg:4326", always_xy=True)

# 좌표 변환 실행
lons, lats = transformer.transform(commercial_gwanak_df['엑스좌표_값'].values, commercial_gwanak_df['와이좌표_값'].values)

commercial_gwanak_df['경도'] = lons
commercial_gwanak_df['위도'] = lats

print("  - 좌표계 변환 완료!")
print("  - 상권 영역 좌표 변환 결과 (상위 5개):")
print(commercial_gwanak_df[['상권_코드_명', '위도', '경도']].head())
print("\n")


# --- 4. 전처리 결과 시각화 ---
# 지오코딩된 쓰레기통과 좌표 변환된 상권 영역을 지도에 표시하여 결과를 확인합니다.
print("--- 4. 전처리 결과 지도 시각화 시작 ---")

# 결측치(좌표 변환 실패)가 없는 데이터만 사용
trash_bins_df_clean = trash_bins_df.dropna(subset=['위도', '경도'])

# 관악구 중심 좌표를 기준으로 지도 생성
map_gwanak = folium.Map(location=[37.478, 126.951], zoom_start=14)

# 지도에 쓰레기통 위치 표시 (파란색 아이콘)
for idx, row in trash_bins_df_clean.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"<b>{row['도로명']}</b><br>{row['설치위치']}",
        icon=folium.Icon(color='blue', icon='trash', prefix='fa')
    ).add_to(map_gwanak)

# 지도에 상권 영역 위치 표시 (빨간색 원)
for idx, row in commercial_gwanak_df.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=5,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.6,
        popup=f"<b>{row['상권_코드_명']}</b>"
    ).add_to(map_gwanak)

# 지도 저장
map_file_path = "preprocessed_data_map.html"
map_gwanak.save(map_file_path)
```

```python
# =================================================================
# 관악구 폐기물 관리 프로젝트 2단계: 데이터 통합 및 특성 공학
# 환경: Google Colab
# 목표: 1. 전처리된 위치 데이터와 유동인구 데이터를 결합
#       2. K-평균 클러스터링에 사용할 최종 분석 데이터셋 생성
#       3. 생성된 데이터를 CSV 파일로 저장하여 다음 단계에 활용
# =================================================================


# --- 1. 유동인구 데이터 업로드 및 불러오기 ---
print("--- 1. 유동인구 데이터 파일 업로드 시작 ---")
from google.colab import files
import io

try:
    print("'서울시_상권분석서비스(길단위인구-상권).csv' 파일을 업로드해주세요.")
    uploaded = files.upload()

    population_filename = '서울시_상권분석서비스(길단위인구-상권).csv'
    # 업로드된 파일 이름을 동적으로 찾기
    for filename in uploaded.keys():
        if '길단위인구' in filename:
            population_filename = filename
            break

    population_df = pd.read_csv(io.BytesIO(uploaded[population_filename]), encoding='cp949')
    print(f"\n  - '{population_filename}' 파일 불러오기 성공!")
    print(f"  - 유동인구 데이터: {len(population_df)} 행\n")

except KeyError:
    print(f"\n[오류] '{population_filename}' 파일을 찾을 수 없습니다.")
    print("파일을 올바르게 선택했는지, 파일 이름이 정확한지 확인해주세요.\n")
    raise
except Exception as e:
    print(f"\n[오류] 파일 처리 중 문제가 발생했습니다: {e}")
    raise

# --- 2. 데이터 통합 (Merge) ---
# 1단계에서 처리한 '관악구 상권 영역' 데이터와 방금 업로드한 '유동인구' 데이터를 결합합니다.
# '상권_코드'를 기준으로 두 데이터를 합쳐, 각 상권의 위치와 유동인구 정보를 함께 봅니다.
print("--- 2. 데이터 통합 시작 (상권 위치 + 유동인구) ---")

# commercial_gwanak_df가 이전 단계에서 생성되어 메모리에 있다고 가정합니다.
# 만약 오류가 발생하면, 이전 단계 코드를 먼저 실행해주세요.
try:
    # 두 데이터프레임의 '상권_코드' 데이터 타입을 통일합니다.
    commercial_gwanak_df['상권_코드'] = commercial_gwanak_df['상권_코드'].astype(int)
    population_df['상권_코드'] = population_df['상권_코드'].astype(int)

    # '상권_코드'를 기준으로 두 데이터프레임을 병합(merge)합니다.
    merged_df = pd.merge(commercial_gwanak_df, population_df, on='상권_코드', how='inner')

    print("  - 데이터 통합 완료!")
    print(f"  - 통합된 데이터: {len(merged_df)} 행")
    # '상권_코드_명_x' -> '상권_코드_명'으로 컬럼 이름 변경
    merged_df.rename(columns={'상권_코드_명_x': '상권_코드_명'}, inplace=True)
    print("  - 통합 결과 (상위 5개):")
    # 주요 컬럼만 선택하여 출력
    print(merged_df[['상권_코드_명', '위도', '경도', '총_유동인구_수']].head())
    print("\n")

except NameError:
    print("\n[오류] 'commercial_gwanak_df' 변수를 찾을 수 없습니다.")
    print("이전 1단계(데이터 전처리) 코드를 먼저 실행하여 변수를 생성해주세요.\n")
    raise


# --- 3. 클러스터링을 위한 최종 데이터셋 준비 ---
# K-평균 클러스터링 모델에 입력으로 사용할 최종 데이터셋을 만듭니다.
# 여기서는 쓰레기 발생 잠재력을 나타내는 '수요 지점'으로 상권 영역들을 사용할 것입니다.
# 클러스터링에 사용할 특성(feature)으로는 위치(위도, 경도)와 총 유동인구 수를 선택합니다.
print("--- 3. K-평균 클러스터링용 최종 데이터셋 생성 ---")

# 클러스터링에 필요한 컬럼만 선택
# '총_유동인구_수'가 0인 경우는 분석에서 제외하여 의미 있는 결과에 집중
features_for_clustering = merged_df[merged_df['총_유동인구_수'] > 0][['위도', '경도', '총_유동인구_수', '상권_코드_명']].copy()

# 결측치 확인 및 처리 (만약 있다면 해당 행 제거)
features_for_clustering.dropna(inplace=True)

print("  - 클러스터링에 사용할 특성: 위도, 경도, 총_유동인구_수")
print(f"  - 최종 분석 대상 수요 지점(상권) 개수: {len(features_for_clustering)} 개")
print("  - 최종 데이터셋 (상위 5개):")
print(features_for_clustering.head())
print("\n")


# --- 4. 최종 데이터 파일로 저장 ---
# 다음 3단계 (K-평균 클러스터링)에서 바로 사용할 수 있도록,
# 지금까지 만든 최종 데이터셋을 CSV 파일로 저장합니다.
output_filename = "gwanak_demand_points_final.csv"
features_for_clustering.to_csv(output_filename, index=False, encoding='utf-8-sig')
```

```python
# --- 1. 유동인구 데이터 업로드 및 불러오기 ---
# '서울시_상권분석서비스(길단위인구-상권).csv' 파일을 업로드
print("--- 1. 유동인구 데이터 파일 업로드 시작 ---")
from google.colab import files
import io

try:
    print("'서울시_상권분석서비스(길단위인구-상권).csv' 파일을 업로드해주세요.")
    uploaded = files.upload()

    population_filename = '서울시_상권분석서비스(길단위인구-상권).csv'
    # 업로드된 파일 이름을 동적으로 찾기
    for filename in uploaded.keys():
        if '길단위인구' in filename:
            population_filename = filename
            break

    population_df = pd.read_csv(io.BytesIO(uploaded[population_filename]), encoding='cp949')
    print(f"  - 유동인구 데이터: {len(population_df)} 행\n")

except KeyError:
    print(f"\n[오류] '{population_filename}' 파일을 찾을 수 없습니다.")
    raise
except Exception as e:
    print(f"\n[오류] 파일 처리 중 문제가 발생했습니다: {e}")
    raise

# --- 2. 데이터 통합 (Merge) ---
# 1단계에서 처리한 '관악구 상권 영역' 데이터와 방금 업로드한 '유동인구' 데이터를 결합합니다.
# '상권_코드'를 기준으로 두 데이터를 합쳐, 각 상권의 위치와 유동인구 정보를 함께 봅니다.
print("--- 2. 데이터 통합 시작 (상권 위치 + 유동인구) ---")

# commercial_gwanak_df가 이전 단계에서 생성되어 메모리에 있다고 가정합니다.
# 만약 오류가 발생하면, 이전 단계 코드를 먼저 실행해주세요.
try:
    # 두 데이터프레임의 '상권_코드' 데이터 타입을 통일합니다.
    commercial_gwanak_df['상권_코드'] = commercial_gwanak_df['상권_코드'].astype(int)
    population_df['상권_코드'] = population_df['상권_코드'].astype(int)

    # '상권_코드'를 기준으로 두 데이터프레임을 병합(merge)합니다.
    merged_df = pd.merge(commercial_gwanak_df, population_df, on='상권_코드', how='inner')

    print("  - 데이터 통합 완료!")
    print(f"  - 통합된 데이터: {len(merged_df)} 행")
    # '상권_코드_명_x' -> '상권_코드_명'으로 컬럼 이름 변경
    merged_df.rename(columns={'상권_코드_명_x': '상권_코드_명'}, inplace=True)
    print("  - 통합 결과 (상위 5개):")
    # 주요 컬럼만 선택하여 출력
    print(merged_df[['상권_코드_명', '위도', '경도', '총_유동인구_수']].head())
    print("\n")

except NameError:
    print("\n[오류] 'commercial_gwanak_df' 변수를 찾을 수 없습니다.")
    print("이전 1단계(데이터 전처리) 코드를 먼저 실행하여 변수를 생성해주세요.\n")
    raise


# --- 3. 클러스터링을 위한 최종 데이터셋 준비 ---
# K-평균 클러스터링 모델에 입력으로 사용할 최종 데이터셋을 만듭니다.
# 유동인구 데이터가 시점별로 중복되므로, 상권별로 그룹화하여 평균 유동인구를 계산합니다.
print("--- 3. K-평균 클러스터링용 최종 데이터셋 생성 ---")

# 상권 코드, 이름, 위치를 기준으로 그룹화하고, 총 유동인구 수의 평균을 계산합니다.
# 이렇게 하면 각 상권이 하나의 데이터 포인트로만 표현됩니다.
final_features_df = merged_df.groupby(['상권_코드', '상권_코드_명', '위도', '경도'])['총_유동인구_수'].mean().reset_index()

# '총_유동인구_수'가 0이거나 결측치인 경우는 분석에서 제외합니다.
final_features_df = final_features_df[final_features_df['총_유동인구_수'] > 0]
final_features_df.dropna(inplace=True)

# 클러스터링에 사용할 최종 특성 선택
features_for_clustering = final_features_df[['위도', '경도', '총_유동인구_수', '상권_코드_명']].copy()

print("  - 클러스터링에 사용할 특성: 위도, 경도, 평균_총_유동인구_수")
print(f"  - 최종 분석 대상 수요 지점(상권) 개수: {len(features_for_clustering)} 개")
print("  - 최종 데이터셋 (상위 5개, 유동인구는 평균값):")
print(features_for_clustering.head())
print("\n")


# --- 4. 최종 데이터 파일로 저장 ---
# 지금까지 만든 최종 데이터셋을 CSV 파일로 저장합니다.
print("--- 4. 최종 데이터셋 파일로 저장 ---")

output_filename = "gwanak_demand_points_final.csv"
features_for_clustering.to_csv(output_filename, index=False, encoding='utf-8-sig')
```

```python
# Colab에서 한글 폰트가 깨지지 않도록 설정
!sudo apt-get install -y fonts-nanum > /dev/null
!sudo fc-cache -fv > /dev/null
!rm ~/.cache/matplotlib -rf > /dev/null
# =================================================================
# 관악구 폐기물 관리 프로젝트 3단계: K-평균 클러스터링 모델링 (2/2)
# 환경: Google Colab
# 목표: 1. 최종 분석 데이터셋을 사용하여 K-평균 클러스터링 수행
#       2. 엘보우 방법을 통해 최적의 클러스터 개수(k) 결정
#       3. 클러스터링 결과를 지도에 시각화하여 최적 입지 제안
# =================================================================


from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import folium
from google.colab import files
import io
import warnings

warnings.filterwarnings('ignore')
# 런타임 재시작 후 폰트를 다시 설정합니다.
plt.rc('font', family='NanumBarunGothic')
print("  - 라이브러리 임포트 및 한글 폰트 설정 완료.\n")


# --- 1. 최종 분석 데이터 불러오기 ---
print("--- 1. 최종 분석 데이터 파일 업로드 시작 ---")
try:
    print("아래 '파일 선택' 버튼을 눌러 'gwanak_demand_points_final.csv' 파일을 업로드해주세요.")
    uploaded = files.upload()

    filename = list(uploaded.keys())[0] # 업로드된 파일의 이름 가져오기

    df = pd.read_csv(io.BytesIO(uploaded[filename]))
    print(f"\n  - '{filename}' 파일 불러오기 성공!")
    print(f"  - 최종 분석 데이터: {len(df)} 행\n")

except Exception as e:
    print(f"\n[오류] 파일 처리 중 문제가 발생했습니다: {e}")
    raise


# --- 2. 클러스터링을 위한 데이터 준비 (특성 선택 및 스케일링) ---
print("--- 2. 클러스터링을 위한 데이터 준비 ---")
# 클러스터링에 사용할 특성: 위도, 경도, 총_유동인구_수
# 여기서는 유동인구 수에 가중치 2를 부여합니다.
features = df[['위도', '경도', '총_유동인구_수']].copy()
features['총_유동인구_수_weighted'] = features['총_유동인구_수'] * 2

# StandardScaler를 사용하여 데이터의 스케일을 표준화합니다.
# 각 특성의 단위와 범위가 다르기 때문에, 표준화는 모델 성능에 매우 중요합니다.
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features[['위도', '경도', '총_유동인구_수_weighted']])
print("  - 특성 선택 및 데이터 스케일링 완료.\n")


# --- 3. 최적의 클러스터 개수(k) 찾기 (엘보우 방법) ---
print("--- 3. 최적의 클러스터 개수(k) 찾기 (엘보우 방법) ---")
inertia = []
K = range(1, 11) # 테스트할 k의 범위 (1~10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# 엘보우 그래프 시각화
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('클러스터 개수 (k)')
plt.ylabel('이너셔 (Inertia)')
plt.title('K-평균 클러스터링을 위한 최적의 k 찾기 (엘보우 방법)')
plt.grid(True)
plt.show()

# 사용자가 그래프를 보고 k를 결정하도록 안내합니다.
print("\n그래프에서 팔꿈치처럼 급격하게 꺾이는 지점이 최적의 k입니다.")
optimal_k = int(input("그래프를 보고 최적의 k 값을 입력하세요 (예: 5): "))
print(f"  - 사용자가 선택한 k 값: {optimal_k}\n")


# --- 4. K-평균 클러스터링 실행 및 결과 분석 ---
print(f"--- 4. k={optimal_k}로 K-평균 클러스터링 실행 ---")
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
# 스케일링된 데이터로 모델 학습
df['cluster'] = kmeans.fit_predict(scaled_features)

# 각 클러스터의 중심점(centroid) 찾기
# 중심점은 스케일링된 공간에서 계산되므로, 이를 다시 원래의 위도/경도/유동인구 스케일로 변환해야 합니다.
centroids_scaled = kmeans.cluster_centers_
centroids = scaler.inverse_transform(centroids_scaled)

# 변환된 중심점을 데이터프레임으로 만듭니다.
centroids_df = pd.DataFrame(centroids, columns=['위도', '경도', '총_유동인구_수_weighted'])
print("  - 클러스터링 완료 및 중심점 계산 완료.")
print("  - 제안되는 최적의 쓰레기통 위치 (클러스터 중심점):")
print(centroids_df[['위도', '경도']])
print("\n")


# --- 5. 클러스터링 결과 시각화 ---
print("--- 5. 클러스터링 결과 지도 시각화 ---")

# 관악구 중심 좌표를 기준으로 지도 생성
map_clusters = folium.Map(location=[37.478, 126.951], zoom_start=14)

# 클러스터별로 색상 지정
# 클러스터 개수가 많아져도 색상이 순환되도록 설정
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'cadetblue', 'darkgreen', 'pink', 'gray']

# 수요 지점(상권)들을 클러스터별로 다른 색으로 표시
for idx, row in df.iterrows():
    cluster_color = colors[row['cluster'] % len(colors)]
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=5,
        color=cluster_color,
        fill=True,
        fill_color=cluster_color,
        fill_opacity=0.7,
        popup=f"<b>{row['상권_코드_명']}</b><br>유동인구: {int(row['총_유동인구_수'])}<br>클러스터: {row['cluster']}"
    ).add_to(map_clusters)

# 클러스터 중심점(최적 위치)을 별 모양 아이콘으로 표시
for i, centroid in centroids_df.iterrows():
    folium.Marker(
        location=[centroid['위도'], centroid['경도']],
        popup=f'<b>추천 위치 {i+1}</b><br>(클러스터 {i} 중심)',
        icon=folium.Icon(color='black', icon='star', prefix='fa')
    ).add_to(map_clusters)

print("  - 지도 생성이 완료되었습니다. 아래에서 결과를 확인하세요.")

# Colab 셀의 마지막에 지도 객체를 두면 자동으로 출력됩니다.
map_clusters
```

```python
# =================================================================
# 관악구 폐기물 관리 프로젝트 4단계
# =================================================================
# --- 라이브러리 설치 및 임포트 ---
print("--- 2단계: 기존 쓰레기통 위치 사전 처리 (1회만 실행) ---")
!pip install geopy > /dev/null
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from google.colab import files
import io, time, re
from tqdm import tqdm

# --- 데이터 업로드 ---
try:
    print("'서울특별시_관악구_가로휴지통_현황_20230905.csv' 파일을 업로드해주세요.")
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    df_raw = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='cp949')
except Exception as e:
    raise Exception(f"파일 처리 중 오류 발생: {e}")

# --- 지오코딩 실행 ---
def create_refined_address(row):
    doro = str(row['도로명']); seolchi = str(row['설치위치'])
    if '동 ' in doro and any(char.isdigit() for char in doro): return f"서울특별시 {doro}"
    doro_cleaned = re.sub(r'([가-힣])(\d)', r'\1 \2', doro).strip()
    if ('로 ' in doro_cleaned or '길 ' in doro_cleaned) and any(char.isdigit() for char in doro_cleaned):
        match = re.search(r'([가-힣\d\s]+(로|길))\s*([\d-]+)', doro_cleaned)
        if match: return f"서울특별시 관악구 {match.group(0)}"
        return f"서울특별시 관악구 {doro_cleaned}"
    return f"서울특별시 관악구 {doro}"
df_raw['refined_address'] = df_raw.apply(create_refined_address, axis=1)

geolocator = Nominatim(user_agent="gwanak-vrp-preprocess")
latitudes, longitudes = [], []
for address in tqdm(df_raw['refined_address'], desc="지오코딩 진행률"):
    try:
        location = geolocator.geocode(address, timeout=10)
        latitudes.append(location.latitude if location else np.nan)
        longitudes.append(location.longitude if location else np.nan)
        time.sleep(1)
    except Exception:
        latitudes.append(np.nan); longitudes.append(np.nan)

df_raw['위도'] = latitudes
df_raw['경도'] = longitudes
df_geocoded = df_raw.dropna(subset=['위도', '경도'])

# --- 결과 파일 저장 및 다운로드 ---
output_filename = 'gwanak_existing_bins_geocoded.csv'
df_geocoded.to_csv(output_filename, index=False, encoding='utf-8-sig')
files.download(output_filename)
print(f"\n  - 지오코딩 완료! '{output_filename}' 파일이 PC에 저장되었습니다.")
print("  - 다음 3단계 코드 셀을 실행할 때 이 파일을 업로드해주세요.")
# =================================================================
# 관악구 폐기물 관리 프로젝트 4단계
# 목표: 최적 수거 경로 탐색 및 시각화
# =================================================================

# --- 0. 라이브러리 설치 및 임포트 ---
print("--- 3단계: 최적 수거 경로 탐색 ---")
!pip install ortools osmnx > /dev/null
print("  - 라이브러리 설치 완료.")
import osmnx as ox
import networkx as nx
from ortools.constraint_solver import routing_enums_pb2
from ortools.constraint_solver import pywrapcp
from geopy.geocoders import Nominatim
from google.colab import files
from tqdm import tqdm
from IPython.display import display

# --- 1. 데이터 업로드 및 위치 정보 정의 ---
print("\n--- 1. 데이터 업로드 및 위치 정보 정의 ---")
try:
    print("'gwanak_existing_bins_geocoded.csv' 파일을 업로드해주세요.")
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    existing_bins_df = pd.read_csv(io.BytesIO(uploaded[filename]))
except Exception as e:
    raise Exception(f"파일 처리 중 오류 발생: {e}")

existing_bins_coords = list(zip(existing_bins_df['위도'], existing_bins_df['경도']))
print(f"  - 기존 쓰레기통 위치 {len(existing_bins_coords)}곳 확인 완료.")

# 차고지(Depot) 위치 설정
try:
    geolocator = Nominatim(user_agent="gwanak-vrp-final-run")
    depot_location = geolocator.geocode("서울특별시 관악구청")
    depot_coords = (depot_location.latitude, depot_location.longitude)
    print(f"  - 차고지 (관악구청) 위치: {depot_coords}")
except Exception:
    depot_coords = (37.4783, 126.9515)

# 신규 추천 위치 좌표
new_bins_coords = [
    (37.485, 126.929), (37.468, 126.943), (37.488, 126.950),
    (37.479, 126.966), (37.498, 126.918)
]
print(f"  - 신규 추천 위치 {len(new_bins_coords)}곳 정의 완료.")

# 전체 위치 리스트 통합
locations = [depot_coords] + new_bins_coords + existing_bins_coords
location_types = ['차고지'] + ['신규 추천'] * len(new_bins_coords) + ['기존 쓰레기통'] * len(existing_bins_coords)
print(f"  - 경로 최적화 대상: 총 {len(locations) - 1}곳\n")

# --- 2. 도로망 데이터 다운로드 및 거리 행렬 생성 ---
print("--- 2. 도로망 데이터 기반 거리 행렬 생성 ---")
G = ox.graph_from_place('Gwanak-gu, Seoul, South Korea', network_type='drive')
nodes = [ox.distance.nearest_nodes(G, Y=lat, X=lon) for lat, lon in locations]
num_locations = len(locations)
dist_matrix = np.zeros((num_locations, num_locations))

for i in tqdm(range(num_locations), desc="거리 행렬 계산 중"):
    for j in range(num_locations):
        if i != j:
            try:
                length = nx.shortest_path_length(G, source=nodes[i], target=nodes[j], weight='length')
                dist_matrix[i][j] = length
            except nx.NetworkXNoPath:
                dist_matrix[i][j] = 9999999 # 경로가 없을 경우 매우 큰 값으로 처리
print("  - 거리 행렬 생성 완료\n")

# --- 3. VRP 문제 설정 및 해결 (Google OR-Tools) ---
print("--- 3. VRP 문제 설정 및 해결 ---")

# [수정] OR-Tools가 정수형 입력을 선호하므로, 거리 행렬을 정수형으로 변환합니다.
dist_matrix_int = dist_matrix.astype(int)

data = {'distance_matrix': dist_matrix_int, 'num_vehicles': 1, 'depot': 0}
manager = pywrapcp.RoutingIndexManager(len(data['distance_matrix']), data['num_vehicles'], data['depot'])
routing = pywrapcp.RoutingModel(manager)

def distance_callback(from_index, to_index):
    from_node = manager.IndexToNode(from_index)
    to_node = manager.IndexToNode(to_index)
    return data['distance_matrix'][from_node][to_node]

transit_callback_index = routing.RegisterTransitCallback(distance_callback)
routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)
search_parameters = pywrapcp.DefaultRoutingSearchParameters()
search_parameters.first_solution_strategy = (routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)
search_parameters.time_limit.FromSeconds(60)
solution = routing.SolveWithParameters(search_parameters)
print("  - 최적 경로 탐색 완료!\n")

# --- 4. 결과 분석 및 시각화 ---
print("--- 4. 최적 경로 결과 분석 및 시각화 ---")
if solution:
    index = routing.Start(0)
    route = [manager.IndexToNode(index)]
    while not routing.IsEnd(index):
        index = solution.Value(routing.NextVar(index))
        route.append(manager.IndexToNode(index))

    route_summary = [location_types[i] if i!=0 else '차고지(출발)' for i in route]
    print(f"  - 최적 경로 (요약): {' -> '.join(route_summary[:6])} ... -> 차고지(도착)")
    print(f"  - 총 이동 거리: {solution.ObjectiveValue() / 1000:.2f} km\n")

    route_map = folium.Map(location=depot_coords, zoom_start=13)
    folium.PolyLine(locations=[locations[i] for i in route], color='blue', weight=4, opacity=0.8).add_to(route_map)

    for i, (loc, loc_type) in enumerate(zip(locations, location_types)):
        if i == 0: icon, popup_text = folium.Icon(color='red', icon='warehouse', prefix='fa'), '<b>차고지</b>'
        elif loc_type == '신규 추천': icon, popup_text = folium.Icon(color='black', icon='star', prefix='fa'), '<b>신규 추천 위치</b>'
        else: icon, popup_text = folium.Icon(color='green', icon='trash', prefix='fa'), '기존 쓰레기통'
        folium.Marker(location=loc, popup=popup_text, icon=icon).add_to(route_map)

    print("  - 최적 경로 지도가 생성되었습니다. 아래에서 확인하세요.")
    display(route_map)
else:
    print('해결책을 찾지 못했습니다.')
```

* 오류 지점 자동 제외: 만약 차량 접근이 불가능한 위치가 발견되면, 해당 지점은 이번 경로 계산에서 자동으로 제외하고 사용자에게 알려줍니다.

```python
# =================================================================
# 관악구 폐기물 관리 프로젝트 4단계
# =================================================================
# --- 라이브러리 설치 및 임포트 ---
!pip install geopy > /dev/null
from geopy.geocoders import Nominatim
from google.colab import files
import io, time, re

# --- 데이터 업로드 ---
try:
    print("'서울특별시_관악구_가로휴지통_현황_20230905.csv' 파일을 업로드해주세요.")
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    df_raw = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='cp949')
    print(f"\n  - '{filename}' 파일 불러오기 성공!")
except Exception as e:
    raise Exception(f"파일 처리 중 오류 발생: {e}")

# --- 지오코딩 실행 ---
def create_refined_address(row):
    doro = str(row['도로명']); seolchi = str(row['설치위치'])
    if '동 ' in doro and any(char.isdigit() for char in doro): return f"서울특별시 {doro}"
    doro_cleaned = re.sub(r'([가-힣])(\d)', r'\1 \2', doro).strip()
    if ('로 ' in doro_cleaned or '길 ' in doro_cleaned) and any(char.isdigit() for char in doro_cleaned):
        match = re.search(r'([가-힣\d\s]+(로|길))\s*([\d-]+)', doro_cleaned)
        if match: return f"서울특별시 관악구 {match.group(0)}"
        return f"서울특별시 관악구 {doro_cleaned}"
    return f"서울특별시 관악구 {doro}"
df_raw['refined_address'] = df_raw.apply(create_refined_address, axis=1)

geolocator = Nominatim(user_agent="gwanak-vrp-preprocess")
latitudes, longitudes = [], []
for address in tqdm(df_raw['refined_address'], desc="지오코딩 진행률"):
    try:
        location = geolocator.geocode(address, timeout=10)
        latitudes.append(location.latitude if location else np.nan)
        longitudes.append(location.longitude if location else np.nan)
        time.sleep(1)
    except Exception:
        latitudes.append(np.nan); longitudes.append(np.nan)

df_raw['위도'] = latitudes
df_raw['경도'] = longitudes
df_geocoded = df_raw.dropna(subset=['위도', '경도'])

# --- 결과 파일 저장 및 다운로드 ---
output_filename = 'gwanak_existing_bins_geocoded.csv'
df_geocoded.to_csv(output_filename, index=False, encoding='utf-8-sig')
files.download(output_filename)
print(f"\n  - 지오코딩 완료. '{output_filename}' 파일이 PC에 저장되었습니다.")
# =================================================================
# 관악구 폐기물 관리 프로젝트 4단계
# 목표: 최적 수거 경로 탐색 및 시각화
# =================================================================

# --- 0. 라이브러리 설치 및 임포트 ---
print("--- 3단계: 최적 수거 경로 탐색 ---")
!pip install ortools osmnx > /dev/null
print("  - 라이브러리 설치 완료.")
import osmnx as ox
import networkx as nx
from ortools.constraint_solver import routing_enums_pb2
from ortools.constraint_solver import pywrapcp
from geopy.geocoders import Nominatim
from IPython.display import display

# --- 1. 데이터 업로드 및 위치 정보 정의 ---
print("\n--- 1. 데이터 업로드 및 위치 정보 정의 ---")
try:
    print("'gwanak_existing_bins_geocoded.csv' 파일을 업로드해주세요.")
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    existing_bins_df = pd.read_csv(io.BytesIO(uploaded[filename]))
    print(f"\n  - '{filename}' 파일 불러오기 성공!")
except Exception as e:
    raise Exception(f"파일 처리 중 오류 발생: {e}")

existing_bins_coords = list(zip(existing_bins_df['위도'], existing_bins_df['경도']))
print(f"  - 기존 쓰레기통 위치 {len(existing_bins_coords)}곳 확인 완료.")

# 차고지(Depot) 위치 설정
try:
    geolocator = Nominatim(user_agent="gwanak-vrp-final-run")
    depot_location = geolocator.geocode("서울특별시 관악구청")
    depot_coords = (depot_location.latitude, depot_location.longitude)
    print(f"  - 차고지 (관악구청) 위치: {depot_coords}")
except Exception:
    depot_coords = (37.4783, 126.9515)

# 신규 추천 위치 좌표
new_bins_coords = [
    (37.485, 126.929), (37.468, 126.943), (37.488, 126.950),
    (37.479, 126.966), (37.498, 126.918)
]
print(f"  - 신규 추천 위치 {len(new_bins_coords)}곳 정의 완료.")

all_locations = [depot_coords] + new_bins_coords + existing_bins_coords
all_location_types = ['차고지'] + ['신규 추천'] * len(new_bins_coords) + ['기존 쓰레기통'] * len(existing_bins_coords)

# --- 2. 도로망 데이터 다운로드 및 정제 ---
print("\n--- 2. 도로망 데이터 다운로드 및 정제 ---")
G = ox.graph_from_place('Gwanak-gu, Seoul, South Korea', network_type='drive')

# [개선] 도로망을 가장 큰 '강한 연결 요소'(Strongly Connected Component)만 남겨 정제합니다.
# 이는 모든 지점 간의 왕복 경로가 보장되는 핵심 도로망만 사용하겠다는 의미입니다.
scc = max(nx.strongly_connected_components(G), key=len)
G_valid = G.subgraph(scc).copy()
print(f"  - 원본 도로망 노드: {len(G.nodes)}, 정제된 핵심 도로망 노드: {len(G_valid.nodes)}")

# [개선] 정제된 도로망에 연결된 유효한 위치만 필터링합니다.
valid_locations, valid_location_types = [], []
print("  - 각 위치의 핵심 도로망 연결성 확인 중...")
for loc, loc_type in zip(all_locations, all_location_types):
    try:
        # 가장 가까운 노드가 G_valid 안에 있는지 확인
        nearest_node = ox.distance.nearest_nodes(G, Y=loc[0], X=loc[1]) # 원본 G에서 찾아야 함
        if nearest_node in G_valid:
            valid_locations.append(loc)
            valid_location_types.append(loc_type)
        else:
            print(f"    [경고] {loc_type} (좌표: {loc})는 핵심 도로망과 연결되지 않아 경로에서 제외합니다.")
    except Exception:
        print(f"    [경고] {loc_type} (좌표: {loc}) 처리 중 오류 발생, 경로에서 제외합니다.")

print(f"\n  - 경로 최적화 대상: 총 {len(valid_locations) - 1}곳 (핵심 도로망에 연결된 위치만 포함)\n")


# --- 3. 거리 행렬 생성 ---
print("--- 3. 거리 행렬 생성 ---")
nodes = [ox.distance.nearest_nodes(G_valid, Y=lat, X=lon) for lat, lon in valid_locations]
num_locations = len(valid_locations)
dist_matrix = np.zeros((num_locations, num_locations))

for i in tqdm(range(num_locations), desc="거리 행렬 계산 중"):
    for j in range(num_locations):
        if i != j:
            length = nx.shortest_path_length(G_valid, source=nodes[i], target=nodes[j], weight='length')
            dist_matrix[i][j] = length
print("  - 거리 행렬 생성!\n")

# --- 4. VRP 문제 설정 및 해결 ---
print("--- 4. VRP 문제 설정 및 해결 ---")
dist_matrix_int = dist_matrix.astype(int)
data = {'distance_matrix': dist_matrix_int, 'num_vehicles': 1, 'depot': 0}
manager = pywrapcp.RoutingIndexManager(len(data['distance_matrix']), data['num_vehicles'], data['depot'])
routing = pywrapcp.RoutingModel(manager)

def distance_callback(from_index, to_index):
    from_node = manager.IndexToNode(from_index)
    to_node = manager.IndexToNode(to_index)
    return data['distance_matrix'][from_node][to_node]

transit_callback_index = routing.RegisterTransitCallback(distance_callback)
routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)
search_parameters = pywrapcp.DefaultRoutingSearchParameters()
search_parameters.first_solution_strategy = (routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)
search_parameters.time_limit.FromSeconds(60)
solution = routing.SolveWithParameters(search_parameters)
print("  - 최적 경로 탐색 완료!\n")

# --- 5. 결과 분석 및 시각화 ---
print("--- 5. 최적 경로 결과 분석 및 시각화 ---")
if solution and solution.ObjectiveValue() > 0:
    index = routing.Start(0)
    route = [manager.IndexToNode(index)]
    while not routing.IsEnd(index):
        index = solution.Value(routing.NextVar(index))
        route.append(manager.IndexToNode(index))

    route_summary = [valid_location_types[i] if i!=0 else '차고지(출발)' for i in route]
    print(f"  - 최적 경로 (요약): {' -> '.join(route_summary[:6])} ... -> 차고지(도착)")
    print(f"  - 총 이동 거리: {solution.ObjectiveValue() / 1000:.2f} km\n")

    route_map = folium.Map(location=depot_coords, zoom_start=13)
    folium.PolyLine(locations=[valid_locations[i] for i in route], color='blue', weight=4, opacity=0.8).add_to(route_map)

    for i, (loc, loc_type) in enumerate(zip(valid_locations, valid_location_types)):
        if i == 0: icon, popup_text = folium.Icon(color='red', icon='warehouse', prefix='fa'), '<b>차고지</b>'
        elif loc_type == '신규 추천': icon, popup_text = folium.Icon(color='black', icon='star', prefix='fa'), '<b>신규 추천 위치</b>'
        else: icon, popup_text = folium.Icon(color='green', icon='trash', prefix='fa'), '기존 쓰레기통'
        folium.Marker(location=loc, popup=popup_text, icon=icon).add_to(route_map)

    # [수정] 지도가 보이지 않는 문제를 해결하기 위해 HTML 파일로 저장하고 다운로드 링크를 제공합니다.
    map_filename = 'gwanak_optimal_route.html'
    route_map.save(map_filename)

    print(f"  - 최적 경로 지도가 '{map_filename}' 파일로 저장되었습니다.")
    files.download(map_filename)
else:
    print('해결책을 찾지 못했습니다. 도로망에 연결된 유효한 지점이 충분한지 확인해주세요.')
```
